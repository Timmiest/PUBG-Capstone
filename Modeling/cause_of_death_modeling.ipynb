#### Here I'm grouping the weapon types to reduce the number of categories and increase the efficacy of the model.

```python
def gun_classifier(damage_type):
    assault_rifles_auto = ['WeapHK416_C','WeapAK47_C', 'WeapBerylM762_C', 'WeapSCAR-L_C']
    assault_rifles_burst = ['WeapMk47Mutant_C','WeapM16A4_C']
    lmg = ['WeapM249_C', 'WeapDP28_C', 'WeapMG3_C']
    pistols = ['WeapM9_C', 'WeapM1911_C', 'WeapG18_C', 'WeapDesertEagle_C', 'Weapvz61Skorpion_C']
    shotguns =['WeapWinchester_C', 'WeapBerreta686_C', 'WeapSaiga12_C','WeapDP12_C']
    sub_machineguns = ['WeapVector_C','WeapUMP_C', 'WeapBizonPP19_C',  'WeapUZI_C','WeapThompson_C', 'WeapNagantM1895_C']
    sniper_dmr = ['WeapVSS_C', 'WeapFNFal_C','WeapMini14_C', 'WeapM24_C','WeapKar98k_C','WeapSKS_C','WeapMosinNagant_C','WeapAWM_C']
    
    if damage_type in assault_rifles_auto:
        return 'assault_rifles_auto'
    if damage_type in assault_rifles_burst:
        return 'assault_rifles_burst'
    if damage_type in lmg:
        return 'lmg'
    if damage_type in pistols:
        return 'pistol'
    if damage_type in shotguns:
        return 'shotgun'
    if damage_type in sub_machineguns:
        return 'sub_machinegun'
    if damage_type in sniper_dmr:
        return 'sniper_dmr'
    else:
        return np.nan
```

```python
wep_predictor_df = df[df['finishDamageInfo_damageTypeCategory'] == 'Damage_Gun'].copy()
```

```python
wep_predictor_df = wep_predictor_df[['death_#','finisher_health','finishDamageInfo_damageCauserName','finishDamageInfo_distance','finishDamageInfo_damageReason','gameState']]
```

```python
wep_predictor_df['weapon_type'] = wep_predictor_df['finishDamageInfo_damageCauserName'].apply(gun_classifier)
```

```python
#removing NA values
wep_predictor_df = wep_predictor_df[wep_predictor_df['weapon_type'].notna()]
```

```python
from sklearn.model_selection import cross_val_score

#modeling fitting function

def model_fitter(model, X_train, y_train, X_test, y_test, cv=5):
    model.fit(X_train, y_train)
    cv_score = np.mean(cross_val_score(model, X_train, y_train, cv=cv))
    training_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    return cv_score, training_score, test_score
```

#### Creating a train/test split, dummifying values and running a range of models

```python
X = wep_predictor_df[['death_#', 'finisher_health',
       'finishDamageInfo_distance', 'finishDamageInfo_damageReason',
       'gameState', 'weapon_type']].copy()
y = X.pop('weapon_type')
X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1, shuffle=True,stratify=y)

scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), 
                       columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), 
                      columns=X_train.columns)

model_list = []

LR = LogisticRegression(max_iter=2000,)
model_list.append(model_fitter(LR, X_train, y_train, X_test, y_test, cv=5))

dtc = DecisionTreeClassifier(max_depth=6)
model_list.append(model_fitter(dtc, X_train, y_train, X_test, y_test, cv=5))

BC = BaggingClassifier()
model_list.append(model_fitter(BC, X_train, y_train, X_test, y_test, cv=5))

knn = KNeighborsClassifier(n_neighbors=5)
model_list.append(model_fitter(knn, X_train, y_train, X_test, y_test, cv=5))

RFC = RandomForestClassifier()
model_list.append(model_fitter(RFC, X_train, y_train, X_test, y_test, cv=5))

LRCV = LogisticRegressionCV(Cs=np.logspace(-4, 4, 10), cv=5, solver='lbfgs')
model_list.append(model_fitter(LRCV, X_train, y_train, X_test, y_test, cv=5))

pd.DataFrame(model_list, columns = ['model','cv score','training score','test score'])
```

#### The logistic regression returns some scores above baseline accuracy. Below, we've run a confusion matrix and examine the coefficients. We predict assault rifles well. We also predict Sniper rifles fairly well. This is explained by the coefficients; with distance and game progress (indicated through the death #) being key indicators. This can be explained somewhat by the viability of certain weapons; a submachine gun is unsuitable for long range kills.
